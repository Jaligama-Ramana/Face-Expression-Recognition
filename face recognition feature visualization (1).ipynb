```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Model
from keras.preprocessing import image
from keras.models import load_model

# Load your custom CNN model
model = load_model('face2.h5')
model.summary()

# Choose a layer to visualize its features
layer_name = 'conv2d_1'  # Change this to the name of the layer you want to visualize
layer_output = model.get_layer(layer_name).output

# Create a new model that outputs the activation values for the chosen layer
activation_model = Model(inputs=model.input, outputs=layer_output)

from PIL import Image

# Load an image to visualize its features
img_path = "C:/Users/Chandhana/Downloads/RAFDB NEW/train/happy/train_09110_aligned.jpg"
img = Image.open(img_path).convert('L')  # Convert to grayscale
img = img.resize((100, 100))  # Resize the image
x = np.expand_dims(np.array(img), axis=0)
x = np.expand_dims(x, axis=-1)  # Add channel dimension

# Get the activations for the chosen layer
activations = activation_model.predict(x)

# Visualize the features
plt.figure(figsize=(16, 8))
for i in range(32):  # Choose the number of filters to visualize
    plt.subplot(1,4, i + 1)
    plt.imshow(activations[0, :, :, i], cmap='viridis')
    plt.axis('off')

plt.show()
```
